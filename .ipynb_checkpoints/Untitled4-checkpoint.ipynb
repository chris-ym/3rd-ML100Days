```python
import pandas as pd
data=[]
with open('c:/python files/hw.txt') as f:
    for line in f:
        line=line.replace('\n',' ').split('\t')
        data.append(line)
```

```python
df=pd.DataFrame(data[1:])
df.columns=data[0]
df_target=df
df_target.index=df.iloc[:,0]
df1=df.iloc[:,1:23]
df1.head()
df2=df.iloc[:,23:]
```

```python
df1_new=df1.astype('str').astype('float')
df2_new=df2.astype('str').astype('float')
```

```python
df1.head()
```

```python
A_group=list(df1.columns)
B_group=list(df2.columns)
```

```python
df_target=df_target.iloc[:,1:]
df_target.head()
```

```python
sum(df1_new.iloc[0,:])/len(df1_new.columns.values.tolist())
```

```python
###行之數值相加
df1_new['mean_1'] = df1_new.apply(lambda x: x.mean(), axis=1)
df1_new['std_1'] = df1_new.apply(lambda x: x.std(), axis=1)

df1_new.index=df.iloc[:,0]
df1_new.head()
```

```python
df2_new['mean_2'] = df2_new.apply(lambda x: x.mean(), axis=1)
df2_new['std_2'] = df2_new.apply(lambda x: x.std(), axis=1)

df2_new.index=df.iloc[:,0]
df2_new.head()
```

```python
df_target['mean_A']=df1_new[['mean_1']]
df_target['mean_B']=df2_new[['mean_2']]

df_target['std_A']=df1_new[['std_1']]
df_target['std_B']=df2_new[['std_2']]


df_target.head()
```

```python
idx=list(df_target.index)
```

```python
from scipy import stats
t,pvalue=stats.ttest_ind(df1_new.loc[idx],df2_new.loc[idx],axis=1)
```

```python
temp_p=pd.DataFrame(pvalue)
temp_p.index=df.iloc[:,0]
temp_p.columns=['pvalue']
```

```python
temp_p.head()
```

```python
df_target['pvalue']=temp_p['pvalue']
```

```python
df_target1=df_target[(df_target['mean_A'])>(df_target['mean_B'])]
df_target1=df_target1.sort_values(['pvalue'])
df_target1.head(10)
```

```python
df_target2=df_target[(df_target['mean_A'])<(df_target['mean_B'])]
df_target2=df_target2.sort_values(['pvalue'])
df_target2.head(10)
```
